```python
eixo_x = videogames["Genre"].value_counts().keys()
eixo_y = videogames["Genre"].value_counts().values
```

```python
plt.bar(eixo_x, eixo_y)

plt.title("Games por generos")

plt.xticks(rotation=45)

plt.xlabel("GÃªneros de jogos")
plt.ylabel("Quantidade de jogos")

plt.show()
```
